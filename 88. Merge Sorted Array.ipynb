Inputs are two sorted list 

Example: 

1 -> 2 -> 4

1 -> 3-> 4
Output:
1->1->2-> 3->4->4

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val 
        self.next = next 

class Solution:
    def MergeSortList(self, l1: ListNode, l2: ListNode) -> ListNode:
        # Create a dummy node:
        dummy = ListNode()
        tail = dummy  # Initialize tail to the dummy node
        
        # Merge the lists:
        while l1 and l2:
            if l1.val < l2.val:
                tail.next = l1  # Add the smaller node to the merged list
                l1 = l1.next 
            else:
                tail.next = l2  # Add the smaller node to the merged list
                l2 = l2.next 
            tail = tail.next  # Update the tail to point to the newly added node
        
        # Handle the remaining nodes:
        if l1:
            tail.next = l1  # Add the remaining nodes from l1
        elif l2:
            tail.next = l2  # Add the remaining nodes from l2
        
        return dummy.next  # Return the merged list
```

```python
# Code execution:

# create a linkedlist:
l1 = ListNode(1)
l1.next = ListNode(2)
l1.next = ListNode(4)

l2 = ListNode(1)
l2.next = ListNode(3)
l2.next = ListNode(4)

# merge the linkedlist>
solution = Solution()
merged_list = solution.MergeSortList(l1, l2)

# print the merged list:

def print_list(head):
    while head:
        print(head.val, end=" ")
        head = head.next
    print()
print_list(merged_list)
```
